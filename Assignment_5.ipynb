#### 4. Compute ngrams  

a. Find the 10 most common bigrams  
b. Find the 10 most common trigrams  

```python
# a) sklearn api? 

import os # I had to import this for the file to create. 

#Clean text 
def text_clean(cleaning_text): 
    cleaning_text= cleaning_text.lower()
    cleaning_text= cleaning_text.replace(","," ")
    cleaning_text= cleaning_text.replace("/", " ")
    cleaning_text= cleaning_text.replace(")", " ")
    cleaning_text= cleaning_text.replace("(", " ")
    
    
    
    return cleaning_text.split()

# Create ngram list and ngram function
def ngrams_list(the_words, num_words): 
    n_list = []
    
    for num in range(0, len(the_words)): 
        ngram= ' '.join(the_words[num:num + num_words])
        n_list.append(ngram)
    
    return n_list


# apply function to text



the_words = text_clean(text)
bigrams = ngrams_list(the_words, 2)



# Save bigrams in separate file because the list is too big for Jupyter Notebook 

b = open('bigrams.txt', mode="w", encoding='utf-8') 
b.write(str(bigrams))

b.close()





    
```

```python
# b) 
trigrams = ngrams_list(the_words, 3)
t = open('trigrams.txt', mode="w", encoding='utf-8')
t.write(str(trigrams))
t.close()

print("Check the assignment folder or where your files from Notebook typically save for the lists!")
```
